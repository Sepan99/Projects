```python
#Creating a new column for grading
food["Letter_grading"] = food["GPA"]

food["Letter_grading"] = pd.to_numeric(food["Letter_grading"], errors="coerce")

food["Letter_grading"] = food["Letter_grading"].fillna(food["Letter_grading"])
```

```python
#Using a function to fill the new column with letter grading
def map_to_letter_grade(score):
    if isinstance(score, str):
        return score
    elif score >= 3.8:
        return "A+"
    elif score >= 3.5:
        return "A"
    elif score >= 3.0:
        return "B+"
    elif score >= 2.7:
        return "B"
    elif score >= 1.7:
        return "C"
    elif score >= 1.0:
        return "D"
    elif score >= 0.0:
        return "F"
    else:
        return None
    
food["Letter_grading"] = food["Letter_grading"].apply(map_to_letter_grade)
```

```python
#Moving the new column
letter_grade = food.pop("Letter_grading")

food.insert(1, "Letter_grading", letter_grade)
```
